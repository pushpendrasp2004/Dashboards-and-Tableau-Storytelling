Q1. Tableau Dashboard vs Story Point
 Tableau Dashboard
A dashboard is a collection of multiple visualizations (charts, graphs, KPIs) displayed on a single screen.
It is used for real-time monitoring and analysis.
 Tableau Story Point
A story is a sequence of dashboards/worksheets arranged to tell a data-driven narrative.
Each step is called a story point.

| Dashboard                       | Story Point                 |
| ------------------------------- | --------------------------- |
| Static/interactive view         | Sequential storytelling     |
| Shows multiple visuals together | Shows insights step-by-step |
| Used for monitoring             | Used for presentation       |

Q2. Actions in Tableau (Filter Action)
- Actions are interactive features in Tableau that allow users to interact with data dynamically.

Filter Action:
It allows one chart to act as a filter for another.
When a user clicks on a value, other charts update accordingly.

Example in Hospital:

Clicking on “West Region” filters:
Only West region patients
Department costs in West
Outcomes in West

 Helps management analyze region-specific performance instantly.
